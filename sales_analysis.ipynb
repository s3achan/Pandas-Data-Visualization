```python
df['Week'] = df['Date'].dt.to_period('W')  # weekly period
df['Week']
```

Challenge - The date is super long in the graph, so resetting the index of my query so that I can get index as 0 - 4/5

```python
weekly = df.groupby(['Week', 'Product'])['Revenue'].sum().unstack().sort_index()
weekly
```

```python
chart1 = weekly.plot(kind='bar', figsize = (9,6), stacked=False,width=0.8,title="Product Revenue by Week",xlabel="Week",ylabel="Revenue")
chart1.set_xticklabels(weekly.index.strftime("Week - %W"))
plt.tight_layout()
plt.show()
```

Product wise revenue - Horizontal Graph


```python
chart2 = weekly.plot(kind='barh', figsize = (9,6), stacked=False,width=0.8,title="Product Revenue by Week",ylabel="Week",xlabel="Revenue")
chart2.set_yticklabels(weekly.index.strftime("Week - %W"))
plt.tight_layout()
plt.show()
```

Creating Individual Charts - One by One

```python

bar_colors = ['tab:purple', 'tab:green', 'tab:orange']

for week, row in weekly.iterrows():
    row.plot.bar(figsize=(5, 3),title=week.strftime("Week - %W"),color=bar_colors)
    plt.ylabel("Revenue")
    plt.tight_layout()
    plt.show()
```

Creating Individual Charts - Same row

```python

fig, axes = plt.subplots(1, len(weekly), figsize=(5 * len(weekly), 4))

for ax, (week, row) in zip(axes, weekly.iterrows()):
    row.plot.bar(ax=ax,color=bar_colors, ylabel="Revenue", xlabel = "Product")
    ax.set_title(week.strftime("Week - %W"))

plt.tight_layout()
plt.show()
```

Creating a multi line chart

```python
chart5 = weekly.plot(kind='line', figsize = (12,6), stacked=False,title="Weekly Revenue Trends by Product",xlabel="Week",ylabel="Revenue")
plt.tight_layout()
plt.show()
```
